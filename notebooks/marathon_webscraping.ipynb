```python
# Chris code for function and loop:

def marathon(url, pages):
    final_list = []
    df = pd.DataFrame()
    for p in range(1, pages + 1):
        response = requests.post(url + str(p))
        soup = BS(response.content, 'lxml')
        tables = soup.find_all('table', attrs = {'class': 'table table-responsive table-bordered'})
        results_list = pd.read_html(str(tables[0]))
        final_list = results_list[0]
        df = df.append(final_list, ignore_index = True)
    return df      
```
